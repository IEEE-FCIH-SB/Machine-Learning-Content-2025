## Lastly Dot product or Matrix multiplication

Key Concepts:

    - Cosine similarity, data representation (embeddings).

    - Daily Use: Recommendation systems, NLP (word2vec).

## Look we agreed before that any AI algorithm or even the computer doesn't understand NL (natural language) as we do as humans 
## so must convert any text to numbers so the algorithm can process it !

### so for now just make sure to understand that we have to convert this text to something called ( vector embeddings ) dw we will discuss it later

```python
# so in the process of comparing the similarity of 2 text we use the dot product to check

# Dot product
a = np.array([1, 2]) # consider this as an embedding for a text ( keda w keda y3ny mtrkzsh awy üòó)
b = np.array([3, 4])
dot_product = np.dot(a, b)  # 11

# or simply u can do this :

dot_product = a @ b

# Cosine similarity
cos_sim = dot_product / (np.linalg.norm(a) * np.linalg.norm(b))  # 0.98

print(cos_sim)
```

## if u are interested on more real life example check this out 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def text_cosine_similarity(text1, text2):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([text1, text2])
    similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
    return similarity

# Example usage
text1 = "ML is bad for your health"
text2 = "AI is good"

similarity_score = text_cosine_similarity(text1, text2)
print(f"Cosine Similarity: {similarity_score:.4f}")
```
