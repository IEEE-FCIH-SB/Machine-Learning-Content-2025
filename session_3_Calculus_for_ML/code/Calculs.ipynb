## 2 - Numerical Differentiation

```python
# Forward Difference Method
def f(x):
    return x**2

x = 3
h = 0.0001
forward_diff = (f(x + h) - f(x)) / h
print(f"Forward Difference Approximation: {forward_diff}")
print(f"Exact Derivative: {2*x}")
```

## So do u think we use numerical diff in ML ? 
### - Well we don't cause it's inefficient and suffers from issues like round-off errors ‚ùå
### - instead in simple problems we use Symbolic Differentiation ‚úÖ 

```python
import sympy as sp

# Define the symbol and function
x = sp.symbols('x')
f = x**2 + 3*x + 2

# Compute the derivative
derivative = sp.diff(f, x)
print(f"Symbolic Derivative: {derivative}")

# Substitute x = 2
value_at_2 = derivative.subs(x, 2)
print(f"Derivative at x = 2 : {value_at_2}")
```

# Auto Diff usning TensorFlow üêçü§ñ
### More efficient for complex problems like deep learning problems 

```python
import tensorflow as tf

# Define a TensorFlow variable
x = tf.Variable(2.0)

# Use TensorFlow's GradientTape for automatic differentiation
with tf.GradientTape() as tape:
    # Define the function y = x^2
    y = x**2 + 3*x + 2

# Compute the derivative dy/dx
dy_dx = tape.gradient(y, x)

print(f"Automatic Derivative: {dy_dx.numpy()}")
```

## F(x) = x**2 try to optimize this function using Gradient Descent and get to the Global minimum point
### - Consider the max iteration or steps is 20 and the initial value of x is -10

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Gradient Descent Parameters
x = -10  # Initial guess
learning_rate = 0.1  # Step size
steps = 20  # Number of iterations

# Store the history of x and y values
x_history = [x]
y_history = [f(x)]

# Gradient Descent Algorithm
for _ in range(steps):
    x = x - learning_rate * df(x)  # Update x
    x_history.append(x)
    y_history.append(f(x))

# Print the final result
print(f"Final x: {x}")
print(f"Final y: {f(x)}")

# Plot the function and the gradient descent steps
x_vals = np.linspace(-11, 11, 400)
plt.plot(x_vals, f(x_vals), label="y = x^2")
plt.scatter(x_history, y_history, color='red', label="Gradient Descent Steps")
plt.legend()
plt.title("Gradient Descent to Minimize y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
