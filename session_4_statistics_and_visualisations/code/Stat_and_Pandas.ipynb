## Histogram (visualize the distribution of numerical data)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# let's create our 1st plot

plt.figure(figsize=(6,4)) # adjusting the figure size 
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black') # customizing the plot
plt.title("Math Score Distribution") # setting the title 
plt.xlabel("Math score") # setting the x-axis label
plt.ylabel("Count") # setting the y-axis label 
plt.show() # Display the plot
```

```python
# alternatively u can use seaborn to do the same 

plt.figure(figsize=(6,4))
sns.histplot(data=df, x='math score' , kde=True)  # Adds a smooth KDE curve
plt.title("Math Score Distribution") # setting the title 
plt.xlabel("Math score") # setting the x-axis label
plt.ylabel("Count") # setting the y-axis label 
plt.show() # Display the plot
```

## Bar Chart (used to compare categorical data using rectangular bars)

- Purpose of Bar Charts
Compare Categories: Show differences between groups (e.g., sales by region).

- Display Counts/Frequencies: Visualize how often categories occur.

- Highlight Trends: Easily spot patterns or outliers in categorical data.

```python
counts = df['gender'].value_counts()


plt.bar(data=df , x= counts.index , height= counts.values , color = ['cornflowerblue' , 'darksalmon'])
plt.xlabel('Gender')
plt.ylabel('Values')
plt.title('Bar Chart for Gender with Matplotlib')
plt.xticks(rotation = 45)
plt.show()
```

```python
# now let's do it with seaborn

sns.barplot(x=counts.index, y=counts.values )
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart for Gender with Seaborn')
plt.xticks(rotation=45)
plt.show()
```

## Count plot (used to visualize the frequency of categorical data)

```python
# Create a bar chart (manual count plot)
plt.bar(counts.index, counts.values, color=['cornflowerblue', 'darkslategray'])
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Count Plot with Matplotlib')
plt.show()
```

```python

# Count plot with seaborn
sns.countplot(data=df, x='gender')
plt.title('Count Plot')
plt.xlabel('Categories')
plt.ylabel('Count with seaborn')
plt.show()
```

## Box plot 
 - (or whisker plot) is a statistical visualization used to summarize the distribution of numerical data. It provides insights into the data's central tendency, spread, and outliers

```python
# look I prefer seaborn over Matplotlib in Box plot tbh but u can use whatever u like 

# Basic Box plot
sns.boxplot(data=df, x = 'math score')
plt.title('Box Plot of math score with Seaborn')
plt.show()
```

```python
# Now let's customize it more !

sns.boxplot(data=df, y='math score' , x = 'gender' , hue='gender' , palette='BuPu')  
plt.title('Box Plot of Math Scores')
plt.ylabel('Math Score')
plt.legend()
plt.show()
```

## Scatter plot (used to visualize the relationship between two numerical variables)

```python
# using matplotlib

# Male plot
plt.scatter(x = 'math score', y = 'reading score', data= df[df['gender'] == 'male'] , color ='skyblue', alpha=0.5, label = 'Male' , s=45)
# female plot
plt.scatter(x = 'math score', y = 'reading score', data= df[df['gender'] == 'female'] , color ='salmon', alpha=0.5, label = 'Female' , s=45)
plt.xlabel('math scores')
plt.ylabel('reading score')
plt.title("student preformance in math & reading")
plt.legend()
plt.show()
```

```python
# now seaborn

sns.scatterplot(data = df , x = 'math score' , y='reading score' , hue = 'gender' , alpha = 0.5 , palette=['salmon','skyblue'] )
plt.title('Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(loc='upper left')
plt.show()
```

## Heat Map (used to visualize the correlation matrix)


```python
# Step 1: Compute the correlation matrix
corr_matrix = df.corr()

# Step 2: Create the heatmap
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## Sub plots 
### so do u see all the plots that we made ? what if we want to put all of that together in the same figure ? ü§î

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns, figure size 12x10

# Plot 1: Histogram of math score
sns.histplot(data=df, x='math score', kde=True, ax=axes[0, 0])  # Top-left
axes[0, 0].set_title('Distribution of Math Scores')

# Plot 2: Box plot of reading score by gender
sns.boxplot(data=df, x='gender', y='reading score', ax=axes[0, 1])  # Top-right
axes[0, 1].set_title('Reading Scores by Gender')

# Plot 3: Scatter plot of math score vs reading score
sns.scatterplot(data=df, x='math score', y='reading score', hue='gender', ax=axes[1, 0])  # Bottom-left
axes[1, 0].set_title('Math Score vs Reading Score')

# Plot 4: Heatmap of correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', ax=axes[1, 1])  # Bottom-right
axes[1, 1].set_title('Correlation Heatmap')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
